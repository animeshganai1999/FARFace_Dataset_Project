```python
import os
import pandas as pd
import shutil
```

```python
def getCommonImages(API_NAME):
    heatmap_folder_path = "heatmap_Male_Female_"+API_NAME
    countries = os.listdir(heatmap_folder_path)
    
    img_info = {}
    
    for country in countries:
        country_path = heatmap_folder_path+'/'+country+'/normal'
        res_types = ['correct','wrong']
        genders = ['female','male']
        for res_type in res_types:
            for gender in genders:
                path = country_path+'/'+res_type+'/'+gender
                all_files = os.listdir(path)
                key = res_type+"_"+gender
#                 print(key)
#                 print(len(all_files))
                try:
                    img_info[key].extend(all_files)
                except:
                    img_info[key] = all_files
    return img_info
```

```python
num_img_categorys = [1,2,5,10]
img_type = ""

img_info_list = []
for num_img_category in num_img_categorys:
    if img_type == "":
        API_NAME = 'result_'+str(num_img_category) #Enter API name
    else:
        API_NAME = 'result_'+img_type+"_"+str(num_img_category) #Enter API name
    img_info = getCommonImages(API_NAME)
    img_info_list.append(img_info)

# keep ony common name in the list for each group like ['correct-male','wrong-male']
correct_male = set(img_info_list[0]['correct_male'])
correct_female = set(img_info_list[0]['correct_female'])
wrong_male = set(img_info_list[0]['wrong_male'])
wrong_female = set(img_info_list[0]['wrong_female'])

for i in range(1,len(num_img_categorys)):
    correct_male_temp = set(img_info_list[i]['correct_male'])
    correct_female_temp = set(img_info_list[i]['correct_female'])
    wrong_male_temp = set(img_info_list[i]['wrong_male'])
    wrong_female_temp = set(img_info_list[i]['wrong_female'])
    
    correct_male = correct_male.intersection(correct_male_temp)
    correct_female = correct_female.intersection(correct_female_temp)
    wrong_male = wrong_male.intersection(wrong_male_temp)
    wrong_female = wrong_female.intersection(wrong_female_temp)
```

```python
def create_folder(path):
    try:
        os.mkdir(path)
    except:
        pass
    
def copyCommonImages(correct_male,correct_female,wrong_male,wrong_female,API_NAME,out_path):
    heatmap_folder_path = "grad-cam-heatmap_"+API_NAME
    
    correct_male_out = out_path+'/'+'correct_male'
    create_folder(correct_male_out)
    
    correct_female_out = out_path+'/'+'correct_female'
    create_folder(correct_female_out)
    
    wrong_male_out = out_path+'/'+'wrong_male'
    create_folder(wrong_male_out)
    
    wrong_female_out = out_path+'/'+'wrong_female'
    create_folder(wrong_female_out)
    
    countries = os.listdir(heatmap_folder_path)
        
    for country in countries:
        country_path = heatmap_folder_path+'/'+country+'/normal'
        all_files = os.listdir(country_path)
        for cor_male in correct_male:
            if cor_male in all_files:
                src = country_path+'/'+cor_male
                dst = correct_male_out+'/'+cor_male
                shutil.copy(src,dst)
                
        for cor_female in correct_female:
            if cor_female in all_files:
                src = country_path+'/'+cor_female
                dst = correct_female_out+'/'+cor_female
                shutil.copy(src,dst)
        
        for wor_male in wrong_male:
            if wor_male in all_files:
                src = country_path+'/'+wor_male
                dst = wrong_male_out+'/'+wor_male
                shutil.copy(src,dst)
        
        for wor_female in wrong_female:
            if wor_female in all_files:
                src = country_path+'/'+wor_female
                dst = wrong_female_out+'/'+wor_female
                shutil.copy(src,dst)
```

```python
for num_img_category in num_img_categorys:
    create_folder('common_images')
    if img_type == "":
        API_NAME = str(num_img_category) #Enter API name
    else:
        API_NAME = img_type+"_"+str(num_img_category) #Enter API name
    create_folder('common_images/result_'+API_NAME)
    copyCommonImages(correct_male,correct_female,wrong_male,wrong_female,API_NAME,'common_images/result_'+API_NAME)
```
