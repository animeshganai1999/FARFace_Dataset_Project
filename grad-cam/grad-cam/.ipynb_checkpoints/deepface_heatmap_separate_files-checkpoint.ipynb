```python
import os
import pandas as pd
import shutil
```

```python
num_img_category = 5
img_type = ""

# if img_type == "":
#     API_NAME = 'result_'+str(num_img_category) #Enter API name
# else:
#     API_NAME = 'result_'+img_type+"_"+str(num_img_category) #Enter API name

API_NAME = "deepface_without_face_detection"

heat_map_path = './../grad-cam-v3/result_original_all_images_wo_face_detection'
res_path = "./../Results in csv files/"+API_NAME

out_path = "heat_map_male_female_all_country_for_all_images_wo_fd"+API_NAME

def createFolder(path):
    try:
        os.mkdir(path)
    except:
        pass
    
createFolder(out_path)
countries = os.listdir(heat_map_path)
for country in countries:
    country_path_heatmap = heat_map_path+"/"+country
    country_path_res = res_path+"/"+country
    out_path_country = out_path+"/"+country
    createFolder(out_path_country)
    
    df_org = pd.read_csv(country_path_res+".csv")
    img_types = os.listdir(country_path_heatmap)
    print("##### "+country+" #####")
    for img_type in img_types:
        print("---- "+img_type+" ----")
        df = df_org.copy()
        valid_type = "valid masked" if img_type == "masked" else "valid normal"
        df = df[['name','gender',img_type,valid_type]]
        df.rename(columns = {valid_type: "valid"}, inplace = True)
        # Remove all the Nan entries
        df = df.dropna()
        img_type_heatmap = country_path_heatmap+"/"+img_type
        out_path_img_type = out_path_country+"/"+img_type
        createFolder(out_path_img_type)
        
        df_correct = df.copy()
        df_wrong = df.copy()
#         print(df)
        df_correct = df_correct[df_correct['gender'] == df_correct[img_type]]
        df_wrong = df_wrong[df_wrong['gender'] != df_wrong[img_type]]
        print("Correct : ",len(df_correct))
        print("Wrong : ",len(df_wrong))
        ## Separate correctly predicted and wrongly predicted images into 2 folders
        # Correctly predicted
        out_path_img_type_correct = out_path_img_type+"/correct"
        createFolder(out_path_img_type_correct)
        out_path_img_type_correct_male = out_path_img_type_correct+'/male'
        createFolder(out_path_img_type_correct_male)
        out_path_img_type_correct_female = out_path_img_type_correct+"/female"
        createFolder(out_path_img_type_correct_female)
        ### Copy Correct Male images
        df_correct_male = df_correct.copy()
        df_correct_male = df_correct_male[df_correct_male['gender'] == 'M']
        fileNames = list(df_correct_male['name'])
        for fileName in fileNames:
            if img_type == "masked":
                fileName = fileName+"_N95"
            fileName = fileName+".jpg"
            try:
                shutil.copy(img_type_heatmap+"/"+fileName,out_path_img_type_correct_male+"/"+fileName)
            except:
                pass
        ### Copy Correct female images
        df_correct_female = df_correct.copy()
        df_correct_female = df_correct_female[df_correct_female['gender'] == 'F']
        fileNames = list(df_correct_female['name'])
        for fileName in fileNames:
            if img_type == "masked":
                fileName = fileName+"_N95"
            fileName = fileName+".jpg"
            try:
                shutil.copy(img_type_heatmap+"/"+fileName,out_path_img_type_correct_female+"/"+fileName)
            except:
                pass
        
        # Wrongly predicted
        out_path_img_type_wrong = out_path_img_type+"/wrong"
        createFolder(out_path_img_type_wrong)
        out_path_img_type_wrong_male = out_path_img_type_wrong+'/male'
        createFolder(out_path_img_type_wrong_male)
        out_path_img_type_wrong_female = out_path_img_type_wrong+'/female'
        createFolder(out_path_img_type_wrong_female)
        
        ### Copy incorrect Male images
        df_wrong_male = df_wrong.copy()
        df_wrong_male = df_wrong_male[df_wrong_male['gender'] == 'M']
        fileNames = list(df_wrong_male['name'])
        for fileName in fileNames:
            if img_type == "masked":
                fileName = fileName+"_N95"
            fileName = fileName+".jpg"
            try:
                shutil.copy(img_type_heatmap+"/"+fileName,out_path_img_type_wrong_male+"/"+fileName)
            except:
                print(img_type_heatmap+"/"+fileName)
                pass
        
        ### Copy incorrect Female images
        df_wrong_female = df_wrong.copy()
        df_wrong_female = df_wrong_female[df_wrong_female['gender'] == 'F']
        fileNames = list(df_wrong_female['name'])
        for fileName in fileNames:
            if img_type == "masked":
                fileName = fileName+"_N95"
            fileName = fileName+".jpg"
            try:
                shutil.copy(img_type_heatmap+"/"+fileName,out_path_img_type_wrong_female+"/"+fileName)
            except:
                print(img_type_heatmap+"/"+fileName)
                pass
```
