## Generate CSV file which containes only 480 test images result (of Original DeepFace)

```python
import os
import pandas as pd
```

```python
def create_folder(path):
    try:
        os.mkdir(path)
    except:
        pass
    
def analysis(result_path,country,df_test,all_columns):    
    df = pd.read_csv(result_path+country+'.csv')
    for i in range(len(df)):
        df.at[i,"name"] = df['name'][i]+"_"+country
    test_data_name = list(df_test['name'])
    df['country'] = country
    df = df[df['name'].isin(test_data_name)]
    df = df[all_columns]
    print(len(df))
    return df
    
img_type = "deepface_ORG" ## trained images type
result_path = "./../Results in csv files/deepface_without_face_detection/"
ground_path = "test_data_trained_on_original/test_data_trained_on_original.csv"
df_test = pd.read_csv(ground_path)
df_test.rename(columns = {'filename':'name'}, inplace = True)
for i in range(len(df_test)):
    df_test.at[i,'name'] = df_test['name'][i]+'_'+df_test['country'][i]
        
out_path = "./result_in_csv_file/"+img_type+"/"
create_folder(out_path)

all_columns = ['name', 'country', 'gender','normal', 'Greyish', 'RGB0.3 Noise', 'RGB0.5 Noise', 'Spread Noise','masked']
df_res = pd.DataFrame(columns = ['name', 'country', 'gender','normal', 'Greyish', 'RGB0.3 Noise', 'RGB0.5 Noise', 'Spread Noise','masked'])

countries = os.listdir(result_path)
for country in countries:
    country = country.split('.')[0]
    df_res = pd.concat([df_res,analysis(result_path,country,df_test,all_columns)])
#     break
df_res = df_res.reset_index(drop = 'index')
df_res.to_csv(out_path+'result.csv',index = False)
```

# Copy into single folder from country wise separated folder

```python
inp_path = "heat_map_male_female_all_country_for_all_images_wo_fd"
out_path = "heatmap_Male_Female_full_dataset_deepface_ORG_wo_fd"

createFolder(out_path)
out_path = out_path+'/'+"original"
createFolder(out_path)

correct_path = out_path+'/correct'
createFolder(correct_path)
correct_male_o = correct_path+'/male'
correct_female_o = correct_path+'/female'
createFolder(correct_male_o)
createFolder(correct_female_o)

wrong_path = out_path+'/wrong'
createFolder(wrong_path)
wrong_male_o = wrong_path+'/male'
wrong_female_o = wrong_path+'/female'
createFolder(wrong_male_o)
createFolder(wrong_female_o)

countries = os.listdir(inp_path)
for country in countries:
    country_path = inp_path+'/'+country+'/'+"normal"
    correct = country_path+'/'+"correct"
    wrong = country_path+'/'+"wrong"
    
    correct_male = correct+'/male'
    correct_female = correct+'/female'
    
    wrong_male = wrong+'/male'
    wrong_female = wrong+'/female'
    
    #copy correct male
    files = os.listdir(correct_male)
    for file in files:
        shutil.copy(correct_male+'/'+file,correct_male_o+'/'+file.split('.')[0]+'_'+country+'.jpg')
    
    #copy wrong male
    files = os.listdir(wrong_male)
    for file in files:
        shutil.copy(wrong_male+'/'+file,wrong_male_o+'/'+file.split('.')[0]+'_'+country+'.jpg')
    
    #copy correct female
    files = os.listdir(correct_female)
    for file in files:
        shutil.copy(correct_female+'/'+file,correct_female_o+'/'+file.split('.')[0]+'_'+country+'.jpg')
        
    #copy wrong female
    files = os.listdir(wrong_female)
    for file in files:
        shutil.copy(wrong_female+'/'+file,wrong_female_o+'/'+file.split('.')[0]+'_'+country+'.jpg')
```
