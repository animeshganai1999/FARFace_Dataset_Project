```python
test_data = pd.read_csv("./../Test Data v3/test_data_trained_on_original/test_data_trained_on_original.csv")
image_status_path = "./../correction/status_csv_files/"
image_path = "./../Players Dataset_v2/original/"
image_path_rgb = "./../Players Dataset_v2/rgb0.3/"
image_details_path = "./../image_details/"
countries = os.listdir(image_status_path)
out_path_train = "tarining_data"
out_path_valid = "validation_data"

def create_folder(path):
    try:
        os.mkdir(path)
    except:
        pass

# The list containing the values "M" and "F"
values_list = ["M", "F"]

# Function to randomly select a value with specified probabilities
def pick_value_with_probability(values_list, probabilities):
    return random.choices(values_list, probabilities)[0]

# Probabilities for selecting each value
probabilities_when_female = [0.85, 0.15] ## {"M","F"}
probabilities_when_male = [0.15, 0.85] ## {"M","F"}

## Get all the male and female images path
male_path = []
female_path = []
for country in countries:
    country = country.split('.')[0]
    image_path_country = image_path+country+"/"
    image_details = pd.read_csv(image_details_path+country+'.csv')
    test_data_country = test_data.copy()
    test_data_country = test_data_country[test_data_country['country'] == country]
    
    ## Use only colored images 
    df_country = pd.read_csv(image_status_path+country+'.csv')
    df_country['filename'] = None
    for i in range(len(df_country)):
        df_country.at[i,'filename'] = df_country['fileName'][i].split('.')[0]
    df_country = df_country[df_country['status'] == 'OK']
    df_country = df_country[['filename']]
    ## Keep only colored images
    image_details = image_details[image_details['filename'].isin(df_country['filename'])]
    
    ## Remove those images which are used in test dataset
    image_details = image_details[~image_details['filename'].isin(test_data_country['filename'])]
    image_details = image_details.reset_index()
    for i in range(len(image_details)):
        if image_details['gender'][i] == 'M':
            male_path.append(image_path_country+image_details['filename'][i]+'.jpg')
        else:
            female_path.append(image_path_country+image_details['filename'][i]+'.jpg')


### Images of training data will be saved be this filename
total_male = len(male_path)
total_female = len(female_path)
cntImg = 0
cntImgVal = 0 
create_folder(out_path_train)
create_folder(out_path_valid)

df_train_gender = pd.DataFrame()
df_train_gender['filename'] = None
df_train_gender['gender'] = None

df_val_gender = pd.DataFrame()
df_val_gender['filename'] = None
df_val_gender['gender'] = None

for country in countries:
    country = country.split('.')[0]
    out_path_country = out_path_train+'/'+country
    ## Original image path
    image_path_country = image_path+country+"/"
    ## RGB 0.3 image path
    image_path_country_rgb = image_path_rgb+country+"/"
    
    df_country = pd.read_csv(image_status_path+country+'.csv')
    image_details = pd.read_csv(image_details_path+country+'.csv')
    test_data_country = test_data.copy()
    test_data_country = test_data_country[test_data_country['country'] == country]
    df_country['filename'] = None
    for i in range(len(df_country)):
        df_country.at[i,'filename'] = df_country['fileName'][i].split('.')[0]
    df_country = df_country[df_country['status'] == 'OK']
    df_country = df_country[['filename']]
    df_res = df_country[~df_country['filename'].isin(test_data_country['filename'])]
    df_res = image_details[image_details['filename'].isin(df_res['filename'])]
    df_res = df_res.reset_index()
    
    ## Keep 70% data for training and rest for validation
    total = len(df_res)
    num_train = ceil(0.7*total)
    num_valid = total-num_train
    all_indices = set(range(0, total + 1))
    df_train_indices = random.sample(range(0, total + 1), num_train)
    df_valid_indices = all_indices - set(df_train_indices)
    
    df_train = df_res[df_res.index.isin(df_train_indices)]
    df_train = df_train.reset_index()
    
    df_valid = df_res[df_res.index.isin(df_valid_indices)]
    df_valid = df_valid.reset_index()
    print(len(df_train))
    print(len(df_valid))

    ######################### Copy training Images ##############################
    out_path_country_anchor = out_path_train+'/anchor/'
    out_path_country_positive = out_path_train+'/positive/'
    out_path_country_negative = out_path_train+'/negative/'
    
    create_folder(out_path_country_anchor)
    create_folder(out_path_country_positive)
    create_folder(out_path_country_negative)

    for i in range(len(df_train)):
        filename = df_train['filename'][i]
        gender = df_train['gender'][i]
        anchor_img = image_path_country+filename+'.jpg'
        pos_img = image_path_country_rgb+filename+'.jpg'
        shutil.copy(anchor_img,out_path_country_anchor+str(cntImg)+'.jpg')
        shutil.copy(pos_img,out_path_country_positive+str(cntImg)+'.jpg')
        
        ## Take negetive images from opposite gender
#         if gender == 'M':
#             neg_img = female_path[random.randint(0,total_female-1)]
#         else:
#             neg_img = male_path[random.randint(0,total_male-1)]
        
        ## Setting-2 (for male images choose female 85% of time and male 15% time and vice versa)
        if gender == 'M':
            selected_value = pick_value_with_probability(values_list, probabilities_when_male)
            if selected_value == "F":
                neg_img = female_path[random.randint(0,total_female-1)]
            else:
                while True:
                    neg_img = male_path[random.randint(0,total_male-1)]
                    if neg_img != anchor_img:
                        break
        else: ## anchor gender is "F"
            selected_value = pick_value_with_probability(values_list, probabilities_when_female)            
            if selected_value == "F":
                while True:
                    neg_img = female_path[random.randint(0,total_female-1)]
                    if neg_img != anchor_img:
                        break
            else:
                neg_img = male_path[random.randint(0,total_male-1)]
            
        shutil.copy(neg_img,out_path_country_negative+str(cntImg)+'.jpg')
        df_train_gender.at[len(df_train_gender.index)] = [str(cntImg),gender]
        cntImg += 1
    
    ######################### Copy validation Images ##############################
    out_path_country_anchor = out_path_valid+'/anchor/'
    out_path_country_positive = out_path_valid+'/positive/'
    out_path_country_negative = out_path_valid+'/negative/'
    
    create_folder(out_path_country_anchor)
    create_folder(out_path_country_positive)
    create_folder(out_path_country_negative)

    for i in range(len(df_valid)):
        filename = df_valid['filename'][i]
        gender = df_valid['gender'][i]
        anchor_img = image_path_country+filename+'.jpg'
        pos_img = image_path_country_rgb+filename+'.jpg'
        shutil.copy(anchor_img,out_path_country_anchor+str(cntImgVal)+'.jpg')
        shutil.copy(pos_img,out_path_country_positive+str(cntImgVal)+'.jpg')
        
#         if gender == 'M':
#             neg_img = female_path[random.randint(0,total_female-1)]
#         else:
#             neg_img = male_path[random.randint(0,total_male-1)]
        
        if gender == 'M':
            selected_value = pick_value_with_probability(values_list, probabilities_when_male)
            if selected_value == "F":
                neg_img = female_path[random.randint(0,total_female-1)]
            else:
                while True:
                    neg_img = male_path[random.randint(0,total_male-1)]
                    if neg_img != anchor_img:
                        break
        else: ## anchor gender is "F"
            selected_value = pick_value_with_probability(values_list, probabilities_when_female)            
            if selected_value == "F":
                while True:
                    neg_img = female_path[random.randint(0,total_female-1)]
                    if neg_img != anchor_img:
                        break
            else:
                neg_img = male_path[random.randint(0,total_male-1)]
                
        shutil.copy(neg_img,out_path_country_negative+str(cntImgVal)+'.jpg')
        df_val_gender.at[len(df_val_gender.index)] = [str(cntImgVal),gender]
        cntImgVal += 1
    print(len(df_res))
#     break

df_train_gender.to_csv('train_gender.csv',index = False)
df_val_gender.to_csv('val_gender.csv',index = False)
```
