```python
import os
import pandas as pd
import json
```

```python
# API_NAME = "facepp" # Enter API name here
# image_type = "rgb0.3"  # Enter Image type

# path = "./../API Result SJ/"+API_NAME+'/'
ground_path = "./../Details/"
```

```python
image_file_path = "./../All Face Images with And Without Mask (N95)/"
def doAnalysis(df,country,image_file_type,facesNotDetected):
    
    image_folder = image_file_path+country+"_resized/"+image_file_type+"/"
    
    #######################################################################################################
    ## All the players with JSON file don't have images (Due to YOLO or MASK some of the images did not generated)
    # to get the correct number of images which are tested in the API are needed
    df['images tested'] = 0
    image_files = os.listdir(image_folder)
    for name in image_files:
        name = name.split('.')[0]
        if image_file_type == "masked":
            name = name.split('_N95')[0]
        rowInd = df.index[df['name'] == name].tolist()[0]
        df.at[rowInd,'images tested'] = 1
        
    df = df[df['images tested'] == 1]
#     print(df)
    ##########################################################################################################
    ## Number of Total Data, Male, Female ##
    total = len(df)
    numMale = len(df[df['gender'] == 'M'])
    numFemale = len(df[df['gender'] == 'F'])
    print("Total number of images : ",total)
    print("Total number of Male images : ",numMale)
    print("Total number of Female images : ",numFemale)
    print("")
    ## Number of faces not detected by API ##
    fails_df = df.copy()
    fails_df = fails_df[df['pgender'].isnull() == True]
    total_not_detected = len(fails_df)
    male_not_detected = len(fails_df[fails_df['gender'] == 'M'])
    female_not_detected = len(fails_df[fails_df['gender'] == 'F'])
    print("Total number of images NOT DETECTED / DETECTED : ",total_not_detected," / ",total - total_not_detected)
    print("Total number of Male images NOT DETECTED / DETECTED : ",male_not_detected," / ",numMale-male_not_detected)
    print("Total number of Female images NOT DETECTED / DETECTED : ",female_not_detected," / ",numFemale-female_not_detected)
    print("")
    ## Overall Gender Accuracy ##
    correctly_predicted = len(df[df['gender'] == df['pgender']])
    accuracy = (correctly_predicted/total)*100
    print("Overall Gender Accuracy : %.2f"%accuracy+"%")
    
    ## Gender Accuracy over Male ##
    df_male = df.copy()
    df_male = df_male[df_male['gender'] == 'M']
    total = len(df_male)
    correctly_predicted = len(df_male[df_male['gender'] == df_male['pgender']])
    accuracy = (correctly_predicted/total)*100
    print("Gender accuracy over Male : %.2f"%accuracy+"%")
    
    ## Gender Accuracy over Female ##
    df_male = df.copy()
    df_male = df_male[df_male['gender'] == 'F']
    total = len(df_male)
    correctly_predicted = len(df_male[df_male['gender'] == df_male['pgender']])
    accuracy = (correctly_predicted/total)*100
    print("Gender accuracy over Female: %.2f"%accuracy+"%")
    print("")
```

```python
def analysis(API_NAME,image_type):
    path = "./../API Result SJ/"+API_NAME+'/'
    countries = os.listdir(path)
    # countries = ['Bangladesh']
    for country in countries:
        file_path = path+country+"/"+image_type
        try:
            # ================ Read the JSON file (Ground Truth of images) =======================
            ground_path_country = ground_path+country
            allFiles = os.listdir(ground_path_country)
            df = pd.DataFrame()
            df['name'] = None
            df['gender'] = None
            for file in allFiles:
                f = open(ground_path_country+'/'+file)
                data = json.load(f)
                name = file.split('.')[0]
                gender = data['gender']
                df.loc[len(df.index)] = [name,gender]
                f.close()
            print("############################## "+country+" ##############################")
        #     print("Ground Truth Data reading DONE")
        #     print(df)
            #====================================================================================

            files = os.listdir(file_path)

            facesNotDetectedNormal = 0
            facesNotDetectedMasked = 0

            for file in files:
                new_path = file_path+"/"+file
                image_file_type = file.split('_')[-1].split('.')[0]
                f = open(new_path,'r')
                lines = f.readlines()

                if image_file_type == "normal":
                    df_normal = df.copy()
                    df_normal['pgender'] = None

                    for line in lines:
                        name = line.split('.')[0].split('_N95')[0]
                        try:
                            gender = line.split(',')[2][0]
                        except:
                            facesNotDetectedNormal += 1
                            continue

        #                 print(name,' ',gender)
                        try:
                            rowInd = df_normal.index[df_normal['name'] == name].tolist()[0]
                            df_normal.at[rowInd,'pgender'] = gender
                        except:
                            continue

                    f.close()
                    if image_type == "original":
                        print("=================== "+image_file_type+" Images ======================")
                    else:
                        print("=================== "+image_type+" Images ======================")
                        
                    doAnalysis(df_normal,country,image_file_type,facesNotDetectedNormal)
                elif image_type == "original":
                    df_masked = df.copy()
                    df_masked['pgender'] = None

                    for line in lines:
                        name = line.split('.')[0].split('_N95')[0]
                        try:
                            gender = line.split(',')[2][0]
                        except:
                            facesNotDetectedMasked += 1
                            continue
        #                 print(name,' ',gender)
                        try:
                            rowInd = df_masked.index[df_masked['name'] == name].tolist()[0]
                            df_masked.at[rowInd,'pgender'] = gender
                        except:
                            continue
                    f.close()
                    print("=================== "+image_file_type+" Images ======================")
                    doAnalysis(df_masked,country,image_file_type,facesNotDetectedMasked)
            print("\n\n")
        except:
            pass
```

```python
analysis(API_NAME = 'facepp',image_type = 'original')
```

```python
analysis(API_NAME = 'facepp',image_type = 'greyish')
```

```python
analysis(API_NAME = 'facepp',image_type = 'rgb0.3')
```

```python
analysis(API_NAME = 'facepp',image_type = 'rgb0.5')
```

```python
analysis(API_NAME = 'facepp',image_type = 'spread')
```
