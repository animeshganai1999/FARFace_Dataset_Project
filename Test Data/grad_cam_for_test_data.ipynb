```python
import os
import pandas as pd
import shutil
```

```python
def getCommonImages(API_NAME):
    heatmap_folder_path = "./../grad-cam/heatmap_Male_Female_"+API_NAME
    countries = os.listdir(heatmap_folder_path)
    
    img_info = {}
    
    for country in countries:
        country_path = heatmap_folder_path+'/'+country+'/normal'
        res_types = ['correct','wrong']
        genders = ['female','male']
        for res_type in res_types:
            for gender in genders:
                path = country_path+'/'+res_type+'/'+gender
                all_files = os.listdir(path)
                key = res_type+"_"+gender
#                 print(key)
#                 print(len(all_files))
                try:
                    img_info[key].extend(all_files)
                except:
                    img_info[key] = all_files
    return img_info
```

```python
num_img_categorys = [1,2,5,10]
img_type = ""

img_info_list = []
for num_img_category in num_img_categorys:
    if img_type == "":
        API_NAME = 'result_'+str(num_img_category) #Enter API name
    else:
        API_NAME = 'result_'+img_type+"_"+str(num_img_category) #Enter API name
    img_info = getCommonImages(API_NAME)
    img_info_list.append(img_info)

# keep only those images which are used for TEST set (480 images)
test_data_path = "test_data_trained_on_"
if img_type == "":
    test_data_path += 'original'
else:
    test_data_path += img_type
test_dataset = os.listdir(test_data_path)

for img_info in img_info_list:
    for key in img_info:
        new_list = []
#         print(img_info[key])
        for data in img_info[key]:
            print(data)
            if data in test_dataset:
                print(data)
                new_list.append(data)
        img_info[key] = new_list
```
